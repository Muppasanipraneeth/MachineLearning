```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("placement.csv")
```

```python
df.head()
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
X=df.iloc[:,0:1]
Y=df.iloc[:-1]
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split

# Example to align lengths (assuming X is longer than Y)
if len(X) > len(Y):
    X = X[:len(Y)]
elif len(Y) > len(X):
    Y = Y[:len(X)]

# Now you can safely split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
m=lr.coef_
c=lr.intercept_
print(m)
```

```python
from sklearn.metrics import mean_squared_error, r2_score,root_mean_squared_error
ypredit=lr.predict(X_test)
mse=mean_squared_error(Y_test,ypredit)
rmse=root_mean_squared_error(Y_test,ypredit)
r2=r2_score(Y_test,ypredit)
```

- Mean square value
- Root mean square value
- R2 score
